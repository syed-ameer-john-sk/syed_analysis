# Finding the number of distinct countries

```python
len(df_new['Country Name'].unique())
```

```python
pd.DataFrame(df_new['Indicator Code'].unique())
```

# Totaling the amount of debt owed by the countries

```python
df_new.groupby('Country Name')['dept'].sum().reset_index()
```

# Country with the highest debt

```python
df_new.groupby('Country Name')['dept'].sum().sort_values(ascending=False).head(1)
```

# Average amount of debt across indicators

```python
df_new.groupby('Country Name')['dept'].agg(np.average).reset_index()
```

# The highest amount of principal repayments

```python
dept_indi=df_new.groupby(['Indicator Code','Indicator Name'])['dept'].sum().reset_index() 
dept_indi.iloc[np.where(dept_indi['Indicator Name'].str.contains('repayments'))].sort_values(by='dept',ascending=False)
```

# The most common debt indicator

```python
df_new['Indicator Code'].value_counts()
```

# Other viable debt issues and conclusion

```python
df_new.groupby('Country Name')['dept'].max().sort_values(ascending=False).reset_index().head(10)
```

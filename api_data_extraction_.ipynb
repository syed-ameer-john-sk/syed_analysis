```python
import pandas as pd
from datetime import date
```

```python
ds=[]
date_generated = pd.date_range(start="2020-01-01",end=date.today())
ds.append(date_generated.strftime("%Y-%m-%d"))
```

```python
ds[0][949]
```

```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
import ssl
import json
import urllib
data=[]
for date in ds[0]:
    
    context = ssl._create_unverified_context()
    with urllib.request.urlopen('https://api.sunrise-sunset.org/json?lat=18.5204&lng=73.8567&date='+date,context=context) as response:
        get= json.loads(response.read().decode('utf-8'))['results']
        a={"date":str(date),"sunrise":str(get['sunrise']),"sunset":str(get['sunset'])}
        data.append(a)
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df2=df.copy()
```

```python
df2['sunrise_utc']=df2['date']+" "+df2['sunrise']
df2['sunset_utc']=df2['date']+" "+df2['sunset']
```

```python
df2['sunrise_utc']=df2['sunrise_utc'].str.replace("AM",'').replace("PM",'').str.strip()
df2['sunset_utc']=df2['sunrise_utc'].str.replace("AM",'').replace("PM",'').str.strip()
```

```python
from datetime import datetime
import datetime as dt
df2['sunrise_utc'] = pd.to_datetime(df2['sunrise_utc'])
df2['sunset_utc'] = pd.to_datetime(df2['sunset_utc'])

df2['sunrise_ist'] = df2['sunrise_utc'].dt.tz_localize('UTC').dt.tz_convert('Asia/Kolkata')
df2['sunset_ist'] = df2['sunset_utc'].dt.tz_localize('UTC').dt.tz_convert('Asia/Kolkata')
```

```python
df2.to_csv('syed.csv')
```

```python
connection = sq.connect('information1.db')
 
# Create a cursor object
curs = connection.cursor()
 
# Run create table sql query
curs.execute("create table if not exists api_Info1" +
             " (date text, sunrise text, sunset text,sunrise_utc text, sunset_utc text,sunrise_ist text,sunset_ist text)")
 
# Load CSV data into Pandas DataFrame
api_data = pd.read_csv('/content/syed.csv')
 
# Write the data to a sqlite db table
api_data.to_sql('api_Info1', connection, if_exists='replace', index=False)
   
# Run select sql query
curs.execute('select * from api_Info1')
 
# Fetch all records
# as list of tuples
records = curs.fetchall()
 
# Display result 
for row in records:
    # show row
    print(row)
     
# Close connection to SQLite database
connection.close()
```
